## 1. แสดงแนวโน้มความสัมพันธ์ของปีและยอดขายเกมทั่วโลกด้วย Regression model

```python
import statsmodels.api as sm
```

```python
# ตัดวิดีโอเกมที่ไม่รู้ปีที่วางขายและปี 2017 ออก (เนื่องจากปี 2017 มีข้อมูลถึงแค่เดือน January)
temp = df[(~df.Year_of_Release.isna()) & (~(df.Year_of_Release.dt.year == 2017))].copy()
temp.Year_of_Release = temp.Year_of_Release.dt.year.astype('int64')
temp = temp.groupby('Year_of_Release')['Global_Sales'].sum().to_frame()
temp = temp.reset_index()
temp
```

```python
X = sm.add_constant(temp.Year_of_Release)
lm = sm.OLS(temp.Global_Sales, X).fit()
lm.params
```

```python
lm.predict([(1,2020), (1,2030), (1, 2040)])
```

```python
sns.lmplot(x='Year_of_Release', y='Global_Sales', data=temp, ci=90, height=6)
```

```python
lm.rsquared
```

### ค่าช่วงความมั่นใจกำหนดไว้ที่ α = 0.1

```python
lm.conf_int(alpha=0.10)
```

```python
lm.pvalues
```

```python
lm.summary()
```

```python
temp.corr(method ='pearson')
```

```python
temp.corr(method ='spearman')
```

## 3. ทดสอบสมมติฐานผลต่างคะแนนวิจารณ์ของนักวิจารณ์และผู้เล่น

คะแนนวิจารณ์ของ Critic มักจะมากกว่าคะแนนวิจารณ์ของ User จริงหรือไม่  
(โดยใช้ข้อมูลที่มีเป็นกลุ่มตัวอย่างเพื่อคาดเดาคำตอบของประชากร)
- H0 : คะแนนวิจารณ์ของ Critic - คะแนนวิจารณ์ของ User <= 0
- H1 : คะแนนวิจารณ์ของ Critic - คะแนนวิจารณ์ของ User > 0  
(right tailed test)  
ที่ระดับนัยสำคัญ α = 0.10

```python
# ตัดแถวที่ไม่ทราบคะแนนวิจารณ์ออก (จำเป็นที่จะต้องรู้คะแนนวิจารณ์ทั้งสองฝั่ง)
score = df[~(df.Critic_Score.isna()) & ~(df.User_Score.isna())]
```

```python
# เนื่องจาก Critic_Score มีช่วงอยู่ที่ 0-100 แต่ User_Score มีช่วงคะแนนอยู่ที่ 0-10 
# จึงทำการ downscale Critic_Score ด้วยการ /= 10
delta = pd.DataFrame(score.Critic_Score/10 - score.User_Score, columns=['score_delta'])
score = pd.concat([score, delta], axis=1)
```

```python
score = score.loc[:, ['Name', 'score_delta']]
score
```

### bootstraping

```python
# สุ่ม 1000 คะแนนวิจารณ์ 1000 ครั้ง
ls = []
for x in range(1000):
    sample = score.score_delta.sample(n=1000, replace=False)
    # concat list
    ls = [*ls, *sample.values]
ls
```

```python
delta = pd.Series(ls)
delta
```

```python
# หาจำนวนครั้งที่ คะแนนวิจารณ์ของ Critic - คะแนนวิจารณ์ของ User <= 0 (คะแนนนักวิจารณ์ น้อยกว่าหรือเท่ากับ คะแนนผู้เล่น)
delta[delta <= 0]
```

```python
# หาค่า p value
sum(delta <= 0) / len(delta)
```

## เขียนสรุปผลการทดสอบสมมติฐาน

เนื่องจาก p value = ~0.59 มากกว่า α = 0.10 จึงสรุปได้ว่ายอมรับ H0 ที่ระดับนัยสำคัญ 0.10  
หมายความว่า คะแนนวิจารณ์ของ Critic - คะแนนวิจารณ์ของ User <= 0 เป็นจริงที่ระดับนัยสำคัญ 0.10  
ซึ่งสามารถสรุปได้ว่า  
- คะแนนวิจารณ์ของนักวิจารณ์นั้นจะ **น้อยกว่าหรือเท่ากับ** คะแนนวิจารณ์ของผู้ใช้งาน (ผู้เล่น) เป็นจริง โดยหากอ้างอิงจากสถานการณ์จริงแล้วอาจเกิดจากได้หลายปัจจัย ตัวอย่างเช่น 
1. นักวิจารณ์จะได้ทดลองเล่นเกมเวอร์ชันที่ยังไม่สมบูรณ์ 100% และยังมีปัญหาอยู่ ซึ่งปัญหาเหล่านั้นส่วนใหญ่จะได้รับการแก้ไขเมื่อทำการวางขายแล้ว ทำให้คะแนนวิจารณ์ของผู้เล่นสูงกว่า
2. นักวิจารณ์ซึ่งเป็นผู้ที่มีประสบการณ์และใช้ทักษะในการวิเคราะห์ตัวเกมหลายแง่มุม อาจะมีเกณฑ์การให้คะแนนที่ยากกว่าซึ่งต่างจากบุคคลทั่วไป
3. ผู้เล่นที่ซื้อเกมส่วนใหญ่มักจะเป็นแฟนเกมหรือสาวกของค่ายพัฒนาเกมนั้น ๆ อยู่แล้ว ส่งผลให้มีตัวแปรด้านอารมณ์ความรู้สึกมาเกี่ยวข้อง ซึ่งอาจเป็นผลให้คะแนนวิจารณ์ของผู้เล่นเพิ่มขึ้นตามไปด้วยได้ ต่างจากนักวิจารณ์ซึ่งมีหน้าที่ในการลดค่าความ bias ให้น้อยที่สุด
